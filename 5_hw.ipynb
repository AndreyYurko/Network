```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
F = 12 * 10 ** 9
u_s = 30 * 10 ** 6
d_i = 2 * 10 ** 6
N = np.array([10, 100, 1000])
```

```python
u = 300 * 10 ** 3
t_1 = [max(n*F/u_s, F/d_i) for n in N]
t_2 = [max(F/u_s, F/d_i, n*F/(u_s + n*u)) for n in N]
plt.plot(N, t_1, marker = 'o')
plt.plot(N, t_2, marker = 'o')
plt.legend(["Клиент-серверная", "Одноранговая"])
plt.show()
```
